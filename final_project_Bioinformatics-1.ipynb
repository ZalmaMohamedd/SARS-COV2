```python
from Bio import motifs
from Bio.Seq import Seq                       
from Bio import SeqIO
import sys
import numpy as np                             #importing all needed libiraries
seqList = []                                   #define array of sequences

inFile = open('delta-alignment.fas','r')       #open fasta file
for record in SeqIO.parse(inFile,'fasta'):     #parse file to read records
    seqList.append(str(record.seq))            #append record to seqlist
seqMat = np.array(seqList)                     #read seq list as numpy array
```

```python
instances = [                                 #read sequences as array of arrays
    Seq(seqMat[0]),
    Seq(seqMat[1]),
    Seq(seqMat[2]),
    Seq(seqMat[3]),
    Seq(seqMat[4]),
    Seq(seqMat[5]),
    Seq(seqMat[6]),
    Seq(seqMat[7]),
    Seq(seqMat[8]),
    Seq(seqMat[9])
]
```

```python
n = len(instances[0])              #define length of array as n (as all arrays have same length)
profile = { 'T':[0]*n,'G':[0]*n ,'C':[0]*n,'A':[0]*n,'N':[0]*n, '-' :[0]*n }     #multiply each char by array lenght

for seq in instances:                #for loop for each seq in array

    for i, char in enumerate(seq):   #iterate over each character
        profile[char][i] += 1



consensus = ""
for i in range(n):
    max_count = 0
    max_nt = 'x'                    #define max nucleotide
    for nt in "ACGTN-":              #loop to compare max count of each nucleotide 
        if profile[nt][i] > max_count:             #if nucleotide  in this index has higher count than max count
            max_count = profile[nt][i]             #set max count to max count of given nucleotide
            max_nt = nt                            
    consensus += max_nt                            #place max nucleotide  in consensus 
print(consensus)
```

```python
from Bio import SeqIO
import re
#find all chemical constituents in consensus sequence
C1 = re.findall(r"C", consensus)              
G1 = re.findall(r"G", consensus)
A1 = re.findall(r"A", consensus)
T1 = re.findall(r"T", consensus)
CG1 = len(C1) + len(G1)

#find average percentages of chemical constituents in consensus sequence  
C_concensus_avg = (len(C1) / len(consensus))*100 
G_concensus_avg = (len(G1) / len(consensus))*100
A_concensus_avg = (len(A1) / len(consensus))*100
T_concensus_avg = (len(T1) / len(consensus))*100
CG_concensus_avg = (CG1 / len(consensus))*100

print(C_concensus_avg)
print(G_concensus_avg)
print(A_concensus_avg)
print(T_concensus_avg)
print(CG_concensus_avg)
```

```python
from Bio import SeqIO
import re

#lists to append contents in
C_case_list = []
G_case_list = []
A_case_list = []
T_case_list = []
CG_case_list = []

for seq_record in SeqIO.parse("alignedOmicron.fas", "fasta"):
    seq = str(seq_record.seq)
    
    #find all chemical constituents in case sequences
    C2 = re.findall(r"C", seq)
    G2 = re.findall(r"G", seq)
    A2 = re.findall(r"A", seq)
    T2 = re.findall(r"T", seq)
    CG2 = len(C2) + len(G2)
    
    #calculate content of each constituent 
    C_count2 = len(C2) / len(seq)
    G_count2 = len(G2) / len(seq)
    A_count2 = len(A2) / len(seq)
    T_count2 = len(T2) / len(seq)
    CG_count2 = CG2 / len(seq)
    
    C_case_list.append(C_count2)
    G_case_list.append(G_count2)
    A_case_list.append(A_count2)
    T_case_list.append(T_count2)
    CG_case_list.append(CG_count2)
    
#find average percentages of chemical constituents in case sequences   
C_case_avg = (sum(C_case_list[0:])/10) * 100
G_case_avg = (sum(G_case_list[0:])/10) * 100
A_case_avg = (sum(A_case_list[0:])/10) * 100
T_case_avg = (sum(T_case_list[0:])/10) * 100
CG_case_avg = (sum(CG_case_list[0:])/10) * 100

print(C_case_avg)
print(G_case_avg)
print(A_case_avg)
print(T_case_avg)
print(CG_case_avg)
```

```python
from Bio import SeqIO
case_list = []              #define array of sequences
for seq_record in SeqIO.parse("omicron_consensus_alignment.fas", "fasta"):      #parse file to read records
    case_list.append(str(seq_record.seq))                                       #append record to case_list

case_mat = np.array(case_list)                                                  #read case_list as numpy array

case_instances = [                                                              #read sequences as array of arrays
    Seq(case_mat[0]),
    Seq(case_mat[1]),
    Seq(case_mat[2]),
    Seq(case_mat[3]),
    Seq(case_mat[4]),
    Seq(case_mat[5]),
    Seq(case_mat[6]),
    Seq(case_mat[7]),
    Seq(case_mat[8]),
    Seq(case_mat[9])
]
```

```python
from Bio.Seq import Seq

n = len(case_instances[0])                                   #define length of array as n (as all arrays have same length)
case = { 'T':[0]*n,'G':[0]*n ,'C':[0]*n,'A':[0]*n,'N':[0]*n, '-' :[0]*n }  #multiply each char by array lenght

for seq in case_instances:                                                #for loop for each seq in array
    for i, char in enumerate(seq):                                        #iterate over each character
        case[char][i] += 1


case_similarity = ""
for i in range(n):
    max_count = 0
    max_nt = ' '                                        #define max nucleotide
    for nt in "ACGTN-":                                 #loop to compare max count of each nucleotide
        if case[nt][i] > max_count:                     #if nucleotide  in this index has higher count than max count
            if case[nt][i] > 7:                         #check if the percentage of the most repeted nucleotied is more than 70%
                max_count = case[nt][i]                 #set max count to max count of given nucleotide
                max_nt = nt                             
    case_similarity += max_nt                           #place max nucleotide in case_similarity
    
new_case_similarity=list(case_similarity)               #convert case_similarity to list 
```
